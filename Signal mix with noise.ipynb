```python
import numpy as np
from scipy import signal as sg
import matplotlib.pyplot as plt
```

```python
time1 = np.linspace(0, 2, 1000)
```

```python
rn3 = np.random.normal(0, 1, len(time1)) # len(time1) makes sure that there are 
```

```python
plt.figure(figsize=(20, 4))
plt.plot(time1, rn3)
```

```python
freq = 21
amp = 20
signal1 = amp*np.sin(2*np.pi*freq*time1)
```

```python
plt.plot(time1, signal1)
```

```python
signal_plus_noise = signal1 + rn3
```

```python
plt.plot(time1, signal_plus_noise)
```
