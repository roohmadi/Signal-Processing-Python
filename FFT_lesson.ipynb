```python
import numpy as np
from scipy import signal as sg
import matplotlib.pyplot as plt
from numpy.fft import fft, ifft
```

```python
#sampling rate
sr = 2000

#sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)

# gelombang pertama
freq1 = 1
amp1 = 3
signal1 = amp1*np.sin(2*np.pi*freq1*t)

# gelombang kedua
freq2 = 4
amp2 = 0.5
signal2 = amp2*np.sin(2*np.pi*freq2*t)

# gelombang ketiga
freq3 = 7
amp3 = 1
signal3 = amp3*np.sin(2*np.pi*freq3*t)

# gelombang campuran/jumlah
signalAll = signal1 + signal2 + signal3 
```

```python
plt.figure(figsize = (32, 6))
plt.subplot(1,3,1)
plt.plot(t, signal1)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(1,3,2)
plt.plot(t, signal2)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(1,3,3)
plt.plot(t, signal3)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
```

```python
plt.figure(figsize = (12, 6))
plt.subplot(1,2,1)
plt.plot(t, signalAll)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

X = fft(signalAll)
N = len(X)
n = np.arange(N)
T = N/sr
freqFFT = n/T

plt.subplot(1,2,2)
plt.stem(freqFFT, np.abs(X),'b', markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0,10)
```
